```python
# prompt: How to change the font size in the above plot

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
# !pip install scipy==1.11.4  # Uncomment if you need to install this specific version
uploaded = files.upload()

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('Data0.csv')


```

```python
# Convert angles to radians
angles_rad = np.deg2rad(data['Orientation'])

# Create a polar stem plot
fig = plt.figure()
ax = fig.add_subplot(111, polar=True)

# Filter data to be within -40 to 40 degrees
filtered_data = data[(data['Orientation'] >= -40) & (data['Orientation'] <= 40)]
filtered_angles_rad = np.deg2rad(filtered_data['Orientation'])

markerline, stemlines, baseline = ax.stem(filtered_angles_rad, filtered_data['Slice'], basefmt=" ")

# Set the color of the stem lines to red
plt.setp(stemlines, 'color', 'g', 'linewidth', 0.5) # Change color and linewidth

# Set the color of the markers to red
plt.setp(markerline, 'markerfacecolor', 'g', 'markeredgecolor', 'g') # Change marker color
plt.setp(markerline, 'markersize', 5)

# Set the theta limits to -40 to 40 degrees
ax.set_thetamin(-40)
ax.set_thetamax(40)

# Set the title and labels (optional)
# ax.set_title('Polar Stem Plot of Cell Alignment Angles (-40 to 40 degrees)')
ax.set_yticklabels([])  # Hide the radial labels
plt.rcParams.update({'font.size': 10})
# Show the plot
plt.show()
```
